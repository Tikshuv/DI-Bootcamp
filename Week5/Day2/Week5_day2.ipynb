```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(50)* 10
y = 3* x + np.random.randn(50)*5

#y = mx+b
coefficients =np.polyfit(x,y,1)
slope, intercept = coefficients

line = slope* x +intercept

plt.scatter(x,y, color ='blue', label= 'scattered points')

plt.plot(x,line, color="red", label="line of best fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
print("slope", slope)
```

```python
from sklearn.linear_model import LinearRegression

# Reshape x for sklearn
X = x.reshape(-1, 1)
print(X)
# Fit the model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot the scatter plot
plt.scatter(x, y, color='blue', label='Scatter Points')

# Plot the regression line
plt.plot(x, y_pred, color='red', label=f'Linear Fit: y={model.coef_[0]:.2f}x+{model.intercept_:.2f}')
plt.title('Scatter Plot with Linear Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Cell 1: Reading and Initial Data Processing
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
customer_data = pd.read_csv('Mall_Customers.csv')

# Create age brackets
customer_data['Age_Bracket'] = pd.cut(customer_data['Age'],
                                    bins=[0, 20, 30, 40, 50, 60, 100],
                                    labels=['<20', '20-30', '31-40', '41-50', '51-60', '60+'])

# Create income brackets
customer_data['Income_Bracket'] = pd.cut(customer_data['Annual_Income_(k$)'],
                                       bins=[0, 30, 60, 90, 120, 150],
                                       labels=['Low', 'Lower-Middle', 'Middle', 'Upper-Middle', 'High'])

# Create spending brackets
customer_data['Spending_Bracket'] = pd.cut(customer_data['Spending_Score'],
                                         bins=[0, 20, 40, 60, 80, 100],
                                         labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

# Display basic information about the dataset
print("Dataset Info:")
print(customer_data.info())
print("\nMissing Values:")
print(customer_data.isnull().sum())
print("\nFirst few rows:")
print(customer_data.head())
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=customer_data,
            x='Annual_Income_(k$)',
            bins=30,
            color='skyblue',
            kde=True)

plt.title('Distribution of Annual Income', fontsize=14, pad=15)
plt.xlabel('Annual Income (k$)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=customer_data,
           x='Age',
           bw_adjust=0.5,
           fill=True,
           color='lightgreen')

plt.title('Age Distribution of Customers', fontsize=14, pad=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.jointplot(data=customer_data,
             x='Annual_Income_(k$)',
             y='Spending_Score',
             kind='hex',
             height=8,
             joint_kws={'gridsize': 20},
             marginal_kws={'bins': 30})

plt.suptitle('Income vs Spending Score Distribution', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=customer_data,
              x='Income_Bracket',
              y='Spending_Score',
              hue='Genre',
              split=True,
              inner='quartile')

plt.title('Spending Score Distribution by Income Bracket and Gender', fontsize=14, pad=15)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.jointplot(data=customer_data,
             x='Age',
             y='Spending_Score',
             kind='reg',
             height=8,
             joint_kws={'scatter_kws': {'alpha':0.5}},
             marginal_kws={'bins': 25})

plt.suptitle('Age vs Spending Score Distribution', y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the figure and subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15))

# Plot 1: Age Distribution
sns.histplot(data=customer_data, x='Age', ax=ax1, stat='density', alpha=0.5)
sns.kdeplot(data=customer_data, x='Age', ax=ax1, color='red')
ax1.set_title('Distribution of Age')
ax1.set_xlabel('Age')

# Plot 2: Income Distribution
sns.histplot(data=customer_data, x='Annual_Income_(k$)', ax=ax2, stat='density', alpha=0.5)
sns.kdeplot(data=customer_data, x='Annual_Income_(k$)', ax=ax2, color='red')
ax2.set_title('Distribution of Annual Income')
ax2.set_xlabel('Annual Income (k$)')

# Plot 3: Spending Distribution
sns.histplot(data=customer_data, x='Spending_Score', ax=ax3, stat='density', alpha=0.5)
sns.kdeplot(data=customer_data, x='Spending_Score', ax=ax3, color='red')
ax3.set_title('Distribution of Spending Score')
ax3.set_xlabel('Spending Score')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()

# If you want to see the distribution of brackets separately,
# we can create count plots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15))

# Plot 1: Age Brackets Distribution
sns.countplot(data=customer_data, x='Age_Bracket', ax=ax1)
ax1.set_title('Distribution of Age Brackets')
ax1.set_xlabel('Age Bracket')
ax1.tick_params(axis='x', rotation=45)

# Plot 2: Income Brackets Distribution
sns.countplot(data=customer_data, x='Income_Bracket', ax=ax2)
ax2.set_title('Distribution of Income Brackets')
ax2.set_xlabel('Income Bracket')
ax2.tick_params(axis='x', rotation=45)

# Plot 3: Spending Brackets Distribution
sns.countplot(data=customer_data, x='Spending_Bracket', ax=ax3)
ax3.set_title('Distribution of Spending Brackets')
ax3.set_xlabel('Spending Bracket')
ax3.tick_params(axis='x', rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Convert Genre to numeric (0 for Female, 1 for Male)
customer_data['Genre_Numeric'] = (customer_data['Genre'] == 'Male').astype(int)

# Create correlation matrix
correlation_matrix = customer_data[['Age', 'Annual_Income_(k$)', 'Spending_Score', 'Genre_Numeric']].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix,
            annot=True,  # Show correlation values
            cmap='coolwarm',  # Color scheme
            vmin=-1, vmax=1,  # Value range
            center=0,  # Center the colormap at 0
            square=True,  # Make cells square
            fmt='.2f',  # Format numbers to 2 decimal places
            linewidths=0.5,  # Add gridlines
            cbar_kws={"shrink": .5})  # Customize colorbar

plt.title('Correlation Matrix of Customer Features', pad=20)
plt.show()
```
