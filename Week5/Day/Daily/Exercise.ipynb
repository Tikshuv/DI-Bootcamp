```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('/content/2019.csv')
df.head()
# df.info()
# df.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Social support'], df['Score'], alpha=0.7, color='blue')
plt.title('Relationship Between Social Support and Happiness Score', fontsize=14)
plt.xlabel('Social Support', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import numpy as np

df_grouped = df.groupby('Country or region')[['GDP per capita', 'Healthy life expectancy']].mean()
fig, ax1 = plt.subplots(figsize=(16, 6))
ax1.bar(df_grouped.index, df_grouped['GDP per capita'], color='b', alpha=0.6, label='GDP per Capita')
ax2 = ax1.twinx()
ax2.plot(df_grouped.index, df_grouped['Healthy life expectancy'], color='r', marker='o', label='Healthy Life Expectancy')
ax1.set_xlabel('Region')
ax1.set_ylabel('GDP per Capita', color='b')
ax2.set_ylabel('Healthy Life Expectancy', color='r')
plt.title('GDP per Capita and Healthy Life Expectancy across Regions')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
