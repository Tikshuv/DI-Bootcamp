```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('/content/global_power_plant_database.csv')
df.head(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['other_fuel1'] = df['other_fuel1'].fillna('Unknown')
df['other_fuel2'] = df['other_fuel2'].fillna('Unknown')
df['other_fuel3'] = df['other_fuel3'].fillna('Unknown')
df['commissioning_year'] = df['commissioning_year'].fillna(df['commissioning_year'].median())
df['owner'] = df['owner'].fillna('Unknown')
df['source'] = df['source'].fillna(df['source'].mode()[0])
df['geolocation_source'] = df['geolocation_source'].fillna('Unknown')
df['url'] = df['url'].fillna('No URL Provided')
df['wepp_id'] = df['wepp_id'].fillna('No WEPP ID')
df['year_of_capacity_data'] = df['year_of_capacity_data'].fillna(df['year_of_capacity_data'].median())
df['generation_data_source'] = df['generation_data_source'].fillna('Unknown')
generation_columns = [col for col in df.columns if 'generation_gwh_' in col]
for col in generation_columns:
    df[col] = df[col].fillna(0)
estimated_generation_columns = [col for col in df.columns if 'estimated_generation_gwh_' in col]
for col in estimated_generation_columns:
    df[col] = df[col].fillna(0)
```

```python
df.isnull().sum()
```

Statistical Analysis:

Perform a statistical analysis of power output by fuel type using NumPy’s statistical functions.
Use hypothesis testing to determine if the mean power output differs significantly between different fuel types.

```python
print('Avg for gwh for 2013: ', np.mean(df['generation_gwh_2013']))
print('stf for ghw for 2013: ', np.std(df['generation_gwh_2013']))
```

```python
from scipy import stats

coal_output = df[df['primary_fuel'] == 'Coal']['generation_gwh_2013']
gas_output = df[df['primary_fuel'] == 'Gas']['generation_gwh_2013']
t_stat, p_value = stats.ttest_ind(coal_output, gas_output)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
# p-value is abysmal hence we can reject null hypothesis
```

Time Series Analysis:

If the dataset includes time-related data (like year of establishment), use NumPy to analyze trends over time.
Explore how the mix of fuel types for power generation has evolved over the years.

```python
yearly_counts = df['commissioning_year'].value_counts().sort_index()
trend = np.polyfit(yearly_counts.index, yearly_counts.values, 1)
print(f"Trend: Slope = {trend[0]}, Intercept = {trend[1]}")
```

Advanced Visualization:

Create visualizations using Matplotlib and Seaborn to illustrate your findings.
Consider plotting the geographical distribution of power plants using latitude and longitude data, if available.

```python
# Power plant distribution by country
country_distribution = df['country'].value_counts()
top_countries = country_distribution.head(20)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Power Plant Distribution by Country (Top 20)', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Power Plants', fontsize=12)
plt.tight_layout()
plt.show()

# Distribution of power plants by fuel type
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='primary_fuel', order=df['primary_fuel'].value_counts().index)
plt.title('Power Plant Distribution by Fuel Type')
plt.xticks(rotation=90)
plt.show()

# Geographical distribution using latitude and longitude
plt.figure(figsize=(16, 8))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='primary_fuel', palette='Set1', alpha=0.7)
plt.title('Geographical Distribution of Power Plants')
# plt.xlabel('Longitude')
# plt.ylabel('Latitude')
plt.axis('off')
plt.show()
```

Matrix Operations in Real-World Context:

Demonstrate matrix operations by analyzing relationships between different attributes (e.g., fuel type, capacity, and geographic location).
Discuss the relevance of eigenvectors and eigenvalues in this context.

```python
from sklearn.decomposition import PCA
numeric_data = df[['capacity_mw', 'latitude', 'longitude']]
pca = PCA(n_components=2)
pca_result = pca.fit_transform(numeric_data)

# Output the eigenvalues (variance explained by each component)
print("Eigenvalues (variance explained):", pca.explained_variance_ratio_)

# Output the principal components (eigenvectors)
print("Principal Components (Eigenvectors):", pca.components_)
```
