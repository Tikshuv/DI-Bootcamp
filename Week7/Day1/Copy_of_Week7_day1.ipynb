```python
from bs4 import BeautifulSoup
import requests

git = 'https://github.com/topics'
response = requests.get(git)
print(response.status_code)
print(response.text[:100])
```

```python
with open('webpage.html', 'w', encoding='utf-8') as f:
    f.write(response.text)
```

```python
import pandas as pd
with open('webpage.html', 'r', encoding='utf-8') as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, 'html.parser')

topic_titles = []
topic_descriptions = []

topic_elements = soup.find_all('p', class_='f3 lh-condensed mb-0 mt-1 Link--primary')
description_elements = soup.find_all('p', class_='f5 color-fg-muted mb-0 mt-1')

for title_element in topic_elements:
    topic_titles.append(title_element.text.strip())

for description_element in description_elements:
    topic_descriptions.append(description_element.text.strip())


print(f"Length of topic titles: {len(topic_titles)}")
print(f"Topic titles: {topic_titles}")

print(f"Length of topic descriptions: {len(topic_descriptions)}")
print(f"Topic descriptions: {topic_descriptions}")


data = {'title': topic_titles, 'description': topic_descriptions}
df = pd.DataFrame(data)
df
```
