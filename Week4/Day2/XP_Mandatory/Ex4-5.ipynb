```python
import pandas as pd
ir = pd.read_csv('/content/Iris.csv')
ir.head()
```

Id is a qualitative as numbers represent labels and don't have any quantitative meaning, they can be replaced with letters and it won't change anything
SepaLengthCM is quantitative as it represents amount of cm
SepalWidthCm is quantitative as it represents amount of cm
PetaLengthCm is quantitative as it represents amount of cm
PetalWidthcm is quantitative as it represents amount of cm
Species is qualitative because it's a category and doesn't have any quantitative meaning

```python
mean_sepal_length = ir['SepalLengthCm'].mean()
print(f"Mean Sepal Length: {mean_sepal_length}")
median_sepal_length = ir['SepalLengthCm'].median()
print(f"Median Sepal Length: {median_sepal_length}")
mode_sepal_length = ir['SepalLengthCm'].mode()[0]
print(f"Mode Sepal Length: {mode_sepal_length}")
```
