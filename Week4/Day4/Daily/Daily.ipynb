```python
import pandas as pd
df = pd.read_csv("/content/Data Science Job Salary dataset/datascience_salaries.csv")
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['normalized_salary'] = scaler.fit_transform(df[['salary']])
df[['salary', 'normalized_salary']].head()
```

```python
from sklearn.decomposition import PCA

numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df[numerical_columns])
df['pca1'] = pca_components[:, 0]
df['pca2'] = pca_components[:, 1]
df[['pca1', 'pca2']].head()
```

```python
salary_stats = df.groupby('experience_level')['salary'].agg(['mean', 'median'])
print(salary_stats)
```
